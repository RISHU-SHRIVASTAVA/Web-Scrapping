```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
page=requests.get(url)
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
mobiles=soup.find_all("div",class_="_1AtVbE col-12-12")
```

```python
print(len(mobiles))
```

```python
for i in mobiles[3:-4]:
    print(i.get_text(),end="\n\n\n")
```

```python
mobiles=soup.find_all("div",class_="_13oc-S")
print(len(mobiles))
```

```python
names=[]
price_list=[]
front_camera=[]
ram=[]
for i in mobiles[3:-4]:
    name=i.find("div",class_='_4rR01T').text
    names.append(name)
    price=i.find("div",class_='_30jeq3 _1_WHN1').text
    price_list.append(price)
    ul=soup.find("ul",class_='_1xgFaf')
    camera=ul.find_all("li")[2].text
    fc=camera.split(" | ")[1].split(" ")[0]
    front_camera.append(fc)
    rams=ul.find_all("li")[0].text
    ram.append(rams)
    
```

```python
df=pd.DataFrame({'Name of the Mobile ':name,"Price":price_list,'Front Camera(MP)':front_camera,"Ram":ram})    
```
